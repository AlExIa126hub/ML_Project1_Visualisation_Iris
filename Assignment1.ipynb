## **The Iris Dataset**: Visualisation

```python
# importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from mpl_toolkits.mplot3d import Axes3D
```

I am importing the Iris dataset from the **Seaborn library** because I think it is better for visualisation (you can clearly see the categories in the species column as they are not encoded like in the version provided by Scikit).

```python
# loading the Iris dataset
iris = sns.load_dataset('iris')
iris.head(10)
```

```python
# inspecting the dataset
iris.info()
```

```python
# getting all unique values of the species column
iris['species'].unique()
```

I will approximate the petal area to be `petal lenght * petal width`. I will name the new feature **petal_area**.

```python
# creating the new feature petal_area
iris['petal_area'] = iris['petal_length'] * iris['petal_width']
```

The newly created column will appear at the end of the dataframe. It would be more relevant to move it together with the other features related to the petal dimension.

```python
# getting the index of the petal_width column
petal_width_index = iris.columns.get_loc('petal_width')

# moving the column petal_area after the column petal_width
iris.insert(petal_width_index + 1, 'petal_area', iris.pop('petal_area'))
```

```python
iris.head(10)
```

I will approximate the sepal area to be `sepal lenght * sepal width`. I will name the new feature **sepal_area**.

```python
# creating the new feature sepal_area
iris['sepal_area'] = iris['sepal_length'] * iris['sepal_width']
```

```python
# getting the index of the sepal_width column
sepal_width_index = iris.columns.get_loc('sepal_width')

# moving the column sepal_area after the column sepal_width
iris.insert(sepal_width_index + 1, 'sepal_area', iris.pop('sepal_area'))
```

```python
iris.head(10)
```

First, I will plot the 4 features of the original dataset (petal length, petal width, sepal length, sepal width) on 3 dimensions.The sepal width feature, although it doesn't have an axis of its own, I will represent it through the dimension of the marker (the bigger the point, the bigger the value). This way the sepal width can be visually compared between the categories.

```python
fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(111, projection = '3d')

for species in iris["species"].unique():
    data = iris[iris["species"] == species]
    ax.scatter3D(data["sepal_length"], data["petal_length"], data["petal_width"], s=data["sepal_width"]*50, label=species, alpha=0.3)

ax.set_xlabel('Sepal Length')
ax.set_ylabel('Petal Length')
ax.set_zlabel('Petal Width')
ax.set_title('3D Scatter Plot by Species with 4 Features')

ax.legend()
plt.show()
```
